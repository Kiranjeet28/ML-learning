```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array(
    [
        [101, 5000, 5200, 5400, 5600],
        [102, 6000, 6100, 6200, 6300],
        [103, 7000, 7100, 7200, 7300],
        [104, 8000, 8200, 8400, 8600],
        [105, 9000, 9100, 9200, 9300],
    ]
)
print("===== Original Sales Data =====")
print("\n Sales data shape", np.shape(sales_data))
print("\n Sample data for 1st 3 restaurants \n", sales_data[0:3, :])


#total sales for each year
total_sales_year = np.sum(sales_data[:, 1:], axis=0)
print("\n Total sales for each year \n", total_sales_year)

# min sales per restaurant
min_sales_restaurant = np.min(sales_data[:, 1:], axis=1)
print("\n Minimum sales per restaurant \n", min_sales_restaurant)

# graph 
plt.figure(figsize=(10, 5))
plt.plot(np.mean(sales_data[:, 1:], axis=1))
plt.title("Total Sales per Year")
plt.xlabel("Years")
```
